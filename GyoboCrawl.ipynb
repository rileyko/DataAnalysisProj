```python
# 데이터 수집
# 1. 교보문고 검색창에 '데이터' 검색
# 2. 창에서 F12를 눌러 마지막 페이지 번호 정보의 tag를 가져와 bs.select로 정보를 가져옴
# 3. 같은 방법으로 책제목, 장르, 단가, 평점, 리뷰수 정보의 tag를 가져옴
# 4. 같은 인덱스끼리 합
# zip()을 이용해 동일한 개수로 이루어진 자료형을 묶어 list()를 통해 리스트로 만듬
# 5. 데이터 프레임에 저장하기 위해 2차원 리스트로 저장
# 6. 데이터프레임으로 변환한 뒤 저장 

import math 
import pandas as pd 
from urllib.request import urlopen
from bs4 import BeautifulSoup

# 도서 검색결과의 전체 페이지 결과를 불러오기 위한 마지막 페이지 번호 계산
html=urlopen('https://search.kyobobook.co.kr/web/search?vPstrKeyWord=%25EB%258D%25B0%25EC%259D%25B4%25ED%2584%25B0&orderClick=LAG')
bs=BeautifulSoup(html,'html.parser')
total=bs.select_one('#searchCategory_1 > small')
pages=int(total.text.replace(',',''))/20
pages=math.ceil(pages) 

total_titles=[]
total_genres=[]
total_prices=[]
total_ratings=[]
total_reviews=[]
total_tags = []
for i in range(1,pages+1):
    print(f'page {i}')
    html=urlopen(f'https://search.kyobobook.co.kr/web/search?vPstrKeyWord=%25EB%258D%25B0%25EC%259D%25B4%25ED%2584%25B0&vPstrTab=PRODUCT&searchPcondition=1&currentPage={i}&orderClick=LAG')
    bs=BeautifulSoup(html,'html.parser')
    title=bs.select('td.detail > div.title > a > strong') #책제목
    genre=bs.select('td.detail > div.title > a > span.category2') #장르
    price=bs.select('td.price > div.org_price > del') #정가
    rating=bs.select('tr > td.info > div.review.klover > a > b') # 평점
    reviews = bs.select('tr > td.info > div.review.booklog > a > b') #리뷰수 
    tags = bs.select('tr > td.detail > div.tag') #태그

    total_titles += title 
    total_genres += genre 
    total_prices += price
    total_ratings += rating 
    total_reviews += reviews
    total_tags +=tags



#같은 인덱스끼리 합치기
books=list(zip(total_titles, total_genres, total_prices, total_ratings, total_reviews, total_tags))
print(len(books))

#데이터프레임에 저장하기 위하여 2차원 리스트로 저장
items=[]
for book in books:
    items.append([book[0].text.strip(),book[1].text.strip().replace('[','').replace(']',''),book[2].text.strip(),book[3].text.strip(),book[4].text.strip(), book[5].text.strip()])

#데이터 프레임으로 변환
df=pd.DataFrame(items, columns=['도서명','카테고리','정가','평점','리뷰수','태그'])

#데이터 프레임 저장
df.to_csv('BookReviews.csv', encoding='ms949')
```
